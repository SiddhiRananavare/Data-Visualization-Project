```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Shree\OneDrive\Desktop\DS Internship\Employee Dataset.csv")
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

### Count Plot

```python
# Count Plot for Groups
plt.figure()
sns.countplot(x='groups', data=df)

plt.title("Count of People in Each Group")
plt.xlabel("Groups")
plt.ylabel("Count")

plt.show()
```

### Histogram

```python
# Age Distribution.

plt.figure()
plt.hist(df['age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Salary Distribution
plt.figure()
plt.hist(df['salary'], bins=10)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

### Bar Chart

```python
# Group-wise Average Salary 
group_salary = df.groupby('groups')['salary'].mean()

plt.figure()
group_salary.plot(kind='bar')
plt.title("Average Salary by Groups")
plt.xlabel("Groups")
plt.ylabel("Average Salary")
plt.show()
```

### Scatter Plot

```python
#  Healthy Eating vs Salary 
plt.figure()
plt.scatter(df['healthy_eating'], df['salary'])
plt.title("Healthy Eating vs Salary")
plt.xlabel("Healthy Eating Score")
plt.ylabel("Salary")
plt.show()
```

### Scatter Plot

```python
#Active Lifestyle vs Salary
plt.figure()
plt.scatter(df['active_lifestyle'], df['salary'])
plt.title("Active Lifestyle vs Salary")
plt.xlabel("Active Lifestyle Score")
plt.ylabel("Salary")
plt.show()
```

### Heatmap

```python
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

### Bar Plot

```python
plt.figure()
sns.boxplot(x='groups', y='salary', data=df)
plt.title("Salary Distribution by Groups")
plt.show()
```

### Pair Plot

```python
sns.pairplot(df)
plt.show()
```
